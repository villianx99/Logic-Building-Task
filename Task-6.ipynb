```python
# Problem 1
def smart_parking_system(capacity, logs):

    current_vehicles = 0
    peak_vehicles = 0

    # Process vehicle logs
    for log in logs:

        if log == "IN":
            current_vehicles += 1

        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_vehicles:
            peak_vehicles = current_vehicles


    # Parking Status Check
    if current_vehicles > capacity:
        status = "Capacity Exceeded "
    else:
        status = "Available"


    return current_vehicles, peak_vehicles, status


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parked, peak, parking_status = smart_parking_system(capacity, vehicle_logs)

print("Currently Parked Vehicles:", parked)
print("Peak Parking Usage:", peak)
print("Parking Status:", parking_status)
```

```python
# Problem 2
def estimate_delivery_time(distance, traffic, weather):

    # Base speed (km per hour)
    base_speed = 30

    # Convert travel time into minutes
    base_time = (distance / base_speed) * 60

    # Traffic delay
    if traffic == "Low":
        base_time += 5
    elif traffic == "Medium":
        base_time += 10
    elif traffic == "High":
        base_time += 20

    # Weather delay
    if weather == "Rainy":
        base_time += 15
    elif weather == "Storm":
        base_time += 25

    return round(base_time)


# Sample Input
distance = 10
traffic = "High"
weather = "Rainy"

eta = estimate_delivery_time(distance, traffic, weather)

print("Estimated Delivery Time:", eta, "minutes")
```

```python
# Problem 3
def analyze_occupancy(total_seats, booked_seats):

    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    if occupancy == 100:
        print("Show Status: Housefull ")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150

analyze_occupancy(total_seats, booked_seats)
```

```python
# Problem 4
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    # Classify load
    if average < 50:
        status = "Normal"
    elif 50 <= average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average, status


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

average, status = classify_server_load(cpu_readings)

print("Average CPU Load:", average, "%")
print("Server Status:", status)
```

```python
# Problem 5
def monitor_resources(resource_usage):
    overused_resources = []

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused_resources.append(resource)

    # Energy alert condition
    if len(overused_resources) > 0:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    return overused_resources, energy_alert


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused, alert = monitor_resources(usage_data)

print("Overused Resources:", ", ".join(overused))
print("Energy Alert:", alert)
```

```python
# Problem 6
def manage_event_registration(capacity, registrations):
    confirmed = 0
    waitlist = 0

    # Loop through registrations
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1

    # Determine event status
    if waitlist > 0:
        status = "Waitlist Activated"
    else:
        status = "Registration Open"

    return confirmed, waitlist, status


# Sample Input
event_capacity = 100
total_registrations = 105

confirmed, waitlist, status = manage_event_registration(event_capacity, total_registrations)

print("Confirmed Registrations:", confirmed)
print("Waitlisted Users:", waitlist)
print("Event Status:", status)
```
