```python
# Problem 1
class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # default state

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def show_status(self):
        print(f"{self.name} is {self.status}")


light_name = input("Light Name: ")
action = input("Action: ")

light = SmartLight(light_name)

if action.upper() == "ON":
    light.turn_on()
elif action.upper() == "OFF":
    light.turn_off()


light.show_status()
```

```python
# Problem 2
class Employee:

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print()
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Taking Input
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

# Creating Object
employee = Employee(name, emp_id, department)

# Displaying ID Card
employee.display_id_card()
```

```python
# Problem 3
class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print()
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Taking Input
name = input("Contact Name: ")
phone = input("Phone Number: ")

# Creating Object
contact1 = Contact(name, phone)

# Displaying Contact
contact1.display_contact()
```

```python
# Problem 4
class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def generate_price_tag(self):
        print(f"Product: {self.name}")
        print()
        print(f"Price: ₹{self.price}")


# Taking Input
name = input("Product Name: ")
price = float(input("Price: "))

# Creating Object
product1 = Product(name, price)

# Displaying Price Tag
product1.generate_price_tag()
```

```python
# Problem 5
class Movie:

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print()
        print(f"Rating: {self.rating} / 5")


# Taking Input
name = input("Movie Name: ")
rating = float(input("Rating: "))

# Creating Object
movie1 = Movie(name, rating)

# Displaying Details
movie1.display_details()
```

```python
# Problem 6
class Delivery:

    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print()
        print(f"Customer: {self.customer_name}")
        print()
        print(f"Address: {self.address}")


# Taking Input
name = input("Customer Name: ")
address = input("Address: ")

# Creating Object
order1 = Delivery(name, address)

# Displaying Details
order1.display_details()
```
